```python
from bs4 import BeautifulSoup
import requests
import csv
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
pages = 3
Names = []
Prices = []

for i in list(range(1,pages+1)):
    print(i)
    url = 'https://www.amazon.in/s?k=Muesli&page={}'.format(str(i))
    print(url)
    source = requests.get('https://www.amazon.in/s?k=Muesli&page=1', headers = headers).text
    soup = BeautifulSoup(source, 'lxml')
    for i in soup.find_all('a', class_='a-link-normal a-text-normal'):
        string = i.text
        Names.append( string.strip() )
    for i in soup.find_all('span', class_='a-price-whole'):
        Prices.append(i.text)


    for i in range(len(Prices)):
        print([i, Names[i], Prices[i]])
        
```

```python
file_name = 'amazonUTF-8.csv'
```

```python
with open(file_name, 'w') as file:
    writer = csv.writer(file)
    writer.writerow(['Sr.No', 'Name', 'Prices'])
    for i in range(len(Prices)):
        writer.writerow([i, Names[i], Prices[i]])
```
