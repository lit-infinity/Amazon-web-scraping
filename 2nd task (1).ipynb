```python
import pandas as pd

brands = pd.read_excel (r"C:\Users\asus\Downloads\New folder\files\brands.xlsx")
print (brands)
```

```python
products = pd.read_excel (r"C:\Users\asus\Downloads\New folder\files\product name.xlsx")

products.columns = [c.replace(' ', '_') for c in products.columns]
print (products.columns)
```

```python
brands = brands['Brands'].to_list()
products = products['Product_Name'].to_list()
```

```python
products = [x.lower() for x in products]
brands = [x.lower() for x in brands]
```

```python
product_list = []
brand_list = []
map = []
for product in products:
    for brand in brands:
        if brand in product:
            product_list.append(product)
            brand_list.append(brand)
        
            map.append(product)
        
```

```python
no_brands = set(products).difference(set(map))
```

```python
print(len(no_brands))
print(len(set(product_list)))
print(len(set(products)))
```
